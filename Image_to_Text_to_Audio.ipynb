```python
# Import libraries
import pytesseract       # For extracting text from image
from PIL import Image    # For image processing
import gtts              # For converting text to speech
```

```python
# Open the image from file path
img = Image.open('/content/Picture1.png')     # Load image from given path
print(img)                                    # Print image details (format, size, mode)
```

```python
# Extract text from the image using OCR
result = pytesseract.image_to_string(img)    # Run OCR and store extracted text
print(result)                                # Print extracted text
```

```python
# Convert the extracted text to speech
tts = gtts.gTTS(result)                     # Generate speech from text
tts.save("hello.mp3")                       # Save the speech to a file
```
